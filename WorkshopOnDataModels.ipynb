```python
def makeNormlizedHistogram(hist_data):
  """
  This function change any histogram into the PDF
  """
  x_val = np.array(hist_data['bucketMeans']) #Get location of the bins
  dx = x_val[1] - x_val[0] # compute bin width
  normalized_hist = np.array(hist_data['histogram']) # Get the histogram
  normalized_hist /= normalized_hist.sum() # make sure that the total sum is one
  normalized_hist /= dx # make sure that the area is one
  return x_val, normalized_hist
```

```python
def imageTo1DArray(im:ee.Image, 
                   aoi:ee.Geometry, 
                   scale:float=10,
                   bands:list=[])->np.ndarray:
  """
  This function extract all the data in ee.image into an numpy array of size
  number of pixels x number of bands.
  """

  imdata = im.reduceRegion(ee.Reducer.toList(), 
                           geometry=aoi, scale=scale).getInfo()
  # extract all pixel data from the image over the AOI
  data = None 
  for band in bands:
    if data is None:
      data = np.array(imdata[band]).reshape(-1,1)
    else: 
      datak = np.array(imdata[band]).reshape(-1,1)
      data = np.concatenate((data, datak), axis=-1)
  return data # shape = num pixels x num bands
```

# Question 5: 

AOI1 and AOI1 are from two types of ground covers which are Rubber and Palm trees, respectively. To be able to distinguish between these two classes, the probability density function (PDF) shuold be different. To measure the differences in the PDFs, we can use the KL-divergences, defined as 

$d_{12} = \frac{1}{2}\left[\log\frac{|\Sigma_1|}{|\Sigma_2|} + \mathrm{tr}\left(\Sigma_2^{-1}\Sigma_1\right) + \left(\mu_1-\mu_2\right)^T\Sigma_2^{-1}\left(\mu_1-\mu_2\right)- N \right]$

where $N$ is the number of bands, $\mu_j$ is the mean vector of AOI $j$, and $\Sigma_j$ is the covariance matrix of AOI $j$.

Thus, the KL distance between two AOIs is defined as

$d_{KL}=\frac{d_{12}+ d_{21}}{2}$

Report $d_{KL}$ of observations in AOI1 and AOI2 on 1-Jan-2021 and 21-Apr-2021 images. Which Dates provide better discrimination power (more $d_{kl}$)?

```python
def kldivergence(mu1:np.ndarray, # mean vector of data 1
                 mu2:np.ndarray, # mean vector of data 2
                 sigma1:np.ndarray, # covariance vector of data 1
                 sigma2:np.ndarray, # covariance vector of data 2
                )->float:
  """
  This function computes the KL divergence between two Gaussian populations. 
  """
  def dij(mui,muj, sigmai,sigmaj): # compute d_ij
    term1 = np.log(np.linalg.det(sigmai)/np.linalg.det(sigmaj))
    term2 = np.trace(np.linalg.solve(sigmaj, sigmai))
    term3 = np.linalg.solve(sigmaj, mu1-mu2)
    term3 = (term3 * (mu1-mu2)).sum()
    distance = 0.5*(term1 + term2 + term3 - len(mui))
    return distance 
  kldist = dij(mu1,mu2,sigma1, sigma2) + dij(mu2,mu1,sigma2, sigma1)
  return 0.5*kldist
```

```python
def addBand3(image):# main function to add one more band to the image
  band1 = image.select('VV')
  band2 = image.select('VH')
  div = band2.divide(band1).rename('VH/VV') # rename new band to VH/VV
  return image.addBands(div).float()
```

```python
mode = 'ASCENDING' # load only ASCENDING MODE
sar_im = ee.ImageCollection('COPERNICUS/S1_GRD_FLOAT') # SENTIEL1 REAL Image 
sar_im = sar_im.filter(ee.Filter.listContains('transmitterReceiverPolarisation', 'VV'))
sar_im = sar_im.filter(ee.Filter.eq('instrumentMode', 'IW'))
sar_im = sar_im.filter(ee.Filter.eq('orbitProperties_pass', mode)) # filter for ASCEDING
dat0 = ee.Date("2020-10-01") # Consider only 1-Oct-2020 to 30-APR-2021
dat1 = ee.Date("2021-04-30")
sar_im = sar_im.filterBounds(aoi3) # only in the aoi3
sar_im = sar_im.filterDate(dat0, dat1)
num_images = sar_im.size().getInfo()
print(num_images)
sar_im = sar_im.sort("system:time_start").select(["VV", "VH"]) # select only VV and VH
sar_im = sar_im.map(addBand3) # add VV/VH
```

6. Run this part to compare the histogram and the Gamma distribution for $\alpha=5$ that is best-fitted with the data.

```python
from numpy.lib.histograms import histogram
import numpy as np 
from datetime import datetime
from scipy.stats import gamma
import matplotlib.pyplot as plt 
sarlist = sar_im.toList(num_images)
plt.figure(figsize=(20,50))
cnt = 1
alpha = 5.0 # SENTINEL 1 perfoms around 5 multilooks
for k in range(0,num_images,1):
  imk = ee.Image(sarlist.get(k))
  date = imk.getInfo()['properties']["system:time_start"]
  date = date_time_obj = datetime.fromtimestamp(date/1000.0) 
  datetxt = date.strftime('%Y-%m-%d')
  print(f"Working on {datetxt}")
  histograms = imk.reduceRegion(ee.Reducer.histogram(100), geometry=aoi3, scale=10).getInfo()
  mean = imk.reduceRegion(ee.Reducer.mean(), geometry=aoi3, scale=10).getInfo()  
  for band in ["VV", "VH", "VH/VV"]:
    x_val, norm_hist = makeNormlizedHistogram(histograms[band])        
    theta = mean[band]/alpha
    plt.subplot(num_images, 3, cnt)
    plt.plot(x_val, norm_hist, label=band)
    if band in ['VV', 'VH']:   
      fx = gamma.pdf(x_val, alpha, scale=theta)
      fx *= norm_hist.sum()/fx.sum() # make sure normalized histogram and pdf have the same area.
      plt.plot(x_val, fx , label="band"+r" $\Gamma$({:0.0f},{:0.3f})".format(alpha, theta))
    plt.title(datetxt)
    if band == "VH":
      plt.xlim(0,0.15)
    else:
      plt.xlim(0,0.6)
    plt.grid()
    plt.legend()
    cnt += 1
```
